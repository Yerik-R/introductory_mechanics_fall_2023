## Definitions

The force generated by static friction, $f_s$, is the force that propels a human forward when the human walks. From the definition of work, the work done by the static friction is:

.

$W_{f_s} = \int_C \vec{f_s} \cdot d\vec{s} = \int_C |\vec{f_s}| |d\vec{s}| \cos{\theta} \;$ ,

.

where $\theta$ is the angle between the force of friction and the displacement. In our model of friction, the angle between the force of friction and the displacement is zero, and $\cos$ of zero degrees is 1. Therefore,

.

$W_{f_s} = \int_C |\vec{f_s}| |d\vec{s}| = \int_C |\mu_s\vec{N}| |d\vec{s}| = \mu_s N \int_C ds = \mu_s N \Delta s \;$ ,

.

where $N$ is the normal force produced by the surface, $\mu_s$ is the coefficient of static friction, and $\Delta s$ is the displacement in the dimension $s$. The *effective* coefficient of static friction assumed by the treadmill is:

$\mu_e = \frac{W_{f_s}}{N \Delta s}$.

## Setting up the problem to include *all* data

When Jorginho runs on the treadmill at an incline, typically the incline is different for different segments of the race. Nevertheless, because work is additive, the equations that describe the work are the form:

.

$mg \left[ \mu_0 \cos\theta_0 s_0 + \mu_1 \cos\theta_1 s_1 + \mu_2 \cos\theta_2 s_2 \right] = W \;$ ,

.

where $s_n$ is the distance run in segment $n$, $\theta_n$ is the angle for segment $n$, and $\mu_n$ is the effective coefficient of friction for segment $n$. For the special case above in which Jorginho's run was completely flat, $s_1 = s_2 = 0$. Moreover, $\theta_0 = 0^{\rm o}$, so $\cos\theta_0 = 1$. Hence, we recover $mg \left[ \mu_0 s_0  \right] = W$, which the equation we used above.

.

10/1/23:  $\; mg \left[ \mu_0 \cos\theta_0 (4.2 {\rm mi}) + \mu_1 \cos\theta_1 (2 {\rm mi}) + \mu_2 \cos\theta_2 (0 {\rm mi}) \right] = 819 {\rm Cal}$

 $ \left[ {\begin{array}{cc}

This matrix equation is similar to the one we saw before, but we have more equations than variables. This system is *overdetermined*, so we can still find its solution using the pseudoinverse:

.


 $ \left[ {\begin{array}{cc}
    \mu_0  \\
    \mu_1 \\
    \mu_2
  \end{array} } \right] =
$
$
  \frac{1}{mg} \left[ {\begin{array}{cc}
    6.2\cos\theta_0 & 0 & 0 \\
    4.2\cos\theta_0 & 2\cos\theta_1 & 0 \\
    2.2\cos\theta_0 & 4\cos\theta_1 & 0 \\
    0.2\cos\theta_0 & 6\cos\theta_1 & 0 \\
    0.2\cos\theta_0 & 4\cos\theta_1 & 2\cos\theta_2
  \end{array} } \right]^{-1}$
$
\left[ {\begin{array}{cc}
    815  \\

    828 \\

  \end{array} } \right]$

````python
# In array form
theta_0 = np.deg2rad(0.0) # These values were computed above
theta_1 = np.deg2rad(2.862405226111748)
theta_2 = np.deg2rad(5.710593137499643)

big_kahuna_array = np.array([[6.2*np.cos(theta_0),                 0.0,               0],
                      [4.2*np.cos(theta_0), 2.0*np.cos(theta_1),               0],
                      [2.2*np.cos(theta_0), 4.0*np.cos(theta_1),               0],
                      [0.2*np.cos(theta_0), 6.0*np.cos(theta_1),               0],
                      [0.2*np.cos(theta_0), 4.0*np.cos(theta_1), 2.0*np.cos(theta_2)]])

big_kahuna_array
````

````python
# convert to SI units (735 is weight in Newtons and 1609 is meters in 1 mile)
big_kahuna_array = 735 * 1609 * big_kahuna_array
big_kahuna_array
````

````python
# Take the pseudoinverse
big_kahuna_array_inv = np.linalg.pinv(big_kahuna_array)
big_kahuna_array_inv
````

````python
# The work array in SI units (1 Calorie is 4184 Jouls)
work_array = np.array([[815],[819],[824],[828],[833]])
work_array = 4184 * work_array
work_array
````

````python
# Just like before, matrix-multiply the inverse times measurement
np.matmul(big_kahuna_array_inv, work_array)
````
